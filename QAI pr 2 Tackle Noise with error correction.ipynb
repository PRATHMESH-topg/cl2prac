```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
def encode_bitflip():
    qc = QuantumCircuit(3)
    qc.barrier()
    qc.cx(0, 1)
    qc.cx(0, 2)
    qc.barrier()
    return qc
```

```python
# Function to simulate a bit-flip error (X gate on qubit 1 for example)
def apply_bitflip_error(qc, qubit):
    qc.x(qubit)
    return qc
```

```python
# Function to decode and correct the error
def decode_bitflip():
    qc = QuantumCircuit(3, 1)
    qc.barrier()
    qc.cx(0, 1)
    qc.cx(0, 2)
    qc.ccx(1, 2, 0)  # Majority vote using Toffoli gate
    qc.barrier()
    qc.measure(0, 0)
    return qc
```

```python
# Create full QEC circuit
qc = QuantumCircuit(3, 1)

```

```python
# Prepare initial state |+> for testing
qc.h(0)

```

```python
# Encode logical qubit
qc = qc.compose(encode_bitflip())
```

```python
# Introduce noise (bit-flip error on qubit 1)
qc = apply_bitflip_error(qc, qubit=1)
```

```python
# Decode and correct
qc = qc.compose(decode_bitflip())
```

```python
from qiskit_aer import Aer
# Simulate
simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()
```

```python
# Show result
print("Measurement Results:", counts)
plot_histogram(counts)
plt.show()
```
